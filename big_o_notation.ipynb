### Big O Notation
- y = 0n + 1 and label the curve "O(1)"
- y = log(n) and label the curve "O(log n)"
- y=n and label the curve "O(n)"
- y=nâˆ—log(n) and label it "O(n log n)"
- y=n^2 and label it "O(n^2)"
- y=2^n and label it "O(2^n)"
- y=n! and label it "O(n!)"
- y=n^n and label it "O(n^n)"


```python
import math
import matplotlib.pyplot as plt
```

```python
x = range(1, 50)
y_1 = [(0 * num) + 1 for num in x]
y_2 = [math.log(num) for num in x]
y_3 = [num for num in x]
y_4 = [num * math.log(num) for num in x]
y_5 = [num ** 2 for num in x]
y_6 = [2 ** num for num in x]
y_7 = [math.factorial(num) for num in x]
y_8 = [num ** num for num in x]
```

- Title your chart "Big O Notation"
- Label your x axis "Elements"
- Label your y axis "Operations"
- Label your curves or make a legend for the curves
- Use LaTex notation where possible

```python
plt.figure(figsize = (10,10))

plt.plot(x, y_1, label='$ \mathcal{O}(1) $', color='green')
plt.plot(x, y_2, label='$ \mathcal{O}(logn) $', color='purple')
plt.plot(x, y_3, label='$ \mathcal{O}(n) $', color='brown')
plt.plot(x, y_4, label='$ \mathcal{O}(nlogn) $', color='red')
plt.plot(x, y_5, label='$ \mathcal{O}(n^{2}) $', color='blue')
plt.plot(x, y_6, label='$ \mathcal{O}(2^{n}) $', color='pink')
plt.plot(x, y_7, label='$ \mathcal{O}(n!) $', color='orange')
plt.plot(x, y_8, label='$ \mathcal{O}(n^{n}) $', color='violet')

plt.xlim(1,20)
plt.ylim(0,100)

plt.legend()
plt.title("Big O Notation")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.show()
```
