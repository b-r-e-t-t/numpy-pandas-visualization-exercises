```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
```

### 1. Use matplotlib to plot the following equation:

  y = x^2 + x + 2
  
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = range(-10, 10)
y = [n**2 - n + 2 for n in x]

plt.plot(x, y)

plt.annotate('Origin',
            xy = (0, 10),
            xytext = (0, 60),
            arrowprops = dict(facecolor = 'gray'))
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

 - y = √x
 - y = x^3
 - y = 2x
 - y = 1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
x_1 = range(0,200)
y_1 = [math.sqrt(num) for num in x_1]
```

```python
plt.plot(x_1, y_1)

plt.title("$ \sqrt{x} $")
```

```python
x_2 = range(0,200)
y_2 = [num ** 3 for num in x_2]
```

```python
plt.plot(x_2, y_2)

plt.title("$ x^3 $")
```

```python
x_3 = range(0,200)
y_3 = [num * 2 for num in x_3]
```

```python
plt.plot(x_3, y_3)

plt.title("$ 2x $")
```

```python
x_4 = range(0,200)
y_4 = [(1/(num + 1)) for num in x_4]
```

```python
plt.plot(x_4, y_4)

plt.title("$ 1/(x+1) $")
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (10,10))

plt.subplot(221)
plt.plot(x_1, y_1)
plt.title("$ \sqrt{x} $")

plt.subplot(222)
plt.plot(x_2, y_2)
plt.title("$ x^3 $")

plt.subplot(223)
plt.plot(x_3, y_3)
plt.title("$ 2x $")

plt.subplot(224)
plt.plot(x_4, y_4)
plt.title("$ 1/(x+1) $")
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (10,10))

plt.plot(x_1, y_1, c = 'red', label = "$ \sqrt{x} $")
plt.plot(x_2, y_2, c = 'purple', label = "$ x^3 $")
plt.plot(x_3, y_3, c = 'green', label = "$ 2x $")
plt.plot(x_4, y_4, c = 'blue', label = "$ 1/(x+1) $")

plt.title("Equations")
plt.legend()
plt.xlim(0, 50)
plt.ylim(0, 50)
plt.show()
```
